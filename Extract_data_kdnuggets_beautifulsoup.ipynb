# Fetching Kdnuggets html page contents

```
import urllib3
html_src = r'https://www.kdnuggets.com/2017/05/42-essential-quotes-data-science-thought-leaders.html'
http = urllib3.PoolManager()
http_resp = http.request('GET',html_src)
kdnuggets_page=http_resp.data

kdnuggets_page
```

# Passing the contents of kdnuggets Page to BeautifulSoup Parser

```
from bs4 import BeautifulSoup as bs
pg1=kdnuggets_page.replace(b'\n',b'')

soup = BeautifulSoup(pg1, 'html.parser')
#pg1
print(soup.prettify())
```

# Fetching Quotes and Authors and putting into a dictionary object

```
soup.find_all(id='post-')
```

```
soup.find_all("ol",attrs={"class":""})
```

```
quo_auth_dict=dict()
all_ols=soup.find_all("ol",attrs={"class":""})
print(len(all_ols))
count=1
for ol in soup.find_all("ol",attrs={"class":""}):
  quote= ol.li.get_text()
  print (count,quote)
  if (ol.next_sibling.a):
    author=ol.next_sibling.a.em.get_text()
    print(author)
  elif (ol.next_sibling.next_sibling.a):
    author=ol.next_sibling.next_sibling.a.em.get_text()
    print(author)
  elif (ol.next_sibling.next_sibling.next_sibling.a):
    author=ol.next_sibling.next_sibling.next_sibling.a.em.get_text()
    print(author)
  else :
    print(count)
  quo_auth_dict.update({quote:author})
  count+=1
  
```
